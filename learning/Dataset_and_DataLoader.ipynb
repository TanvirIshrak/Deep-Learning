```python
from sklearn.datasets import make_classification
import torch
```

```python
x,y = make_classification(
    n_samples=10,
    n_features= 2,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    random_state=42
)
```

```python
# Custom dataset

from torch.utils.data import Dataset, DataLoader

class CustomDataSet(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels
        
    def __len__(self):
        return self.features.shape[0]  # returns length
    
    def __getitem__(self, index):
        return self.features[index], self.labels[index]
```

```python
dataset = CustomDataSet(x,y)
```

```python
len(dataset)
```

##### Data Loader

```python
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)  # solving mini batching problem
```
