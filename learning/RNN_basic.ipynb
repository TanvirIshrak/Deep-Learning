```python
import torch
import torch.nn as nn
```

```python
class RNNModel(nn.Module):
    def __init__(self):
        #register kora
        super(RNNModel, self).__init__()
        
        self.rnn = nn.RNN(input_size=5, hidden_size=3, batch_first=True)
        self.fc = nn.Linear(3,1)  #fc = fully connected layer
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        # x e sequence pathai
        out,_ = self.rnn(x)   # _ -> hidden , RNN returns output & hidden together
        out = out[:,-1:]
        out = self.fc(out)
        out = self.sigmoid(out)
        return out
        
```

```python
model = RNNModel()
model
```

```python
# Count trainable parameters
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(model)

```

```python
print(f"\nTrainable parameters breakdown:")

for name, param in model.named_parameters():
    if param.requires_grad:
        print(f"  {name}: {param.numel()} params {list(param.shape)}")

print(f"\nTotal Trainable Parameters: {total_params}")
```
