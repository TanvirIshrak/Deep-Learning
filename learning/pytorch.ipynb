## Reduction operation

```python
e = torch.randint(size=(2,3), low=0 , high=10, dtype=torch.float32)
e
```

```python
torch.sum(e)
```

```python
torch.mean(e)
```

```python
# column wise sum
torch.sum(e , axis=0) 
# torch.sum(e , dim=0) 
```

```python
torch.max(e)
```

```python
torch.prod(e)
```

```python
torch.argmax(e)
```

## Matrix operation

```python
f = torch.randint(size=(2,3), low=0, high=10)
g = torch.randint(size=(3,2), low=0, high=10)
```

```python
torch.matmul(f,g)
```

```python
# dot product
vector1 = torch.tensor([1,2])
vector2 = torch.tensor([3,4])

torch.dot(vector1,vector2)
```

```python
torch.transpose(f, 0,1)
```

```python
h = torch.randint(size=(3,3), low=0 , high=10, dtype=torch.float32)
```

```python
# determinent
torch.det(h)
```

```python
torch.inverse(h)
```
