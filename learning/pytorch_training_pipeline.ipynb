```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head(20)
```

```python
df.drop(columns=['id','Unnamed: 32'], inplace=True)
```

```python
x = df.iloc[: , 1:]
y = df.iloc[: , 0]

x_train, x_test, y_train, y_test  = train_test_split(
    x,y,test_size=0.2, random_state=42
)
```

```python
df.info()
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)
```

```python
type(y_test_encoded)
```

### Numpy array to tensors conversion

```python
x_train_scaled_tensor = torch.from_numpy(x_train_scaled)
x_test_scaled_tensor = torch.from_numpy(x_test_scaled)
y_train_encoded_tensor = torch.from_numpy(y_train_encoded)
x_test_encoded_tensor = torch.from_numpy(y_test_encoded)
```

```python
type(x_train_scaled_tensor)
```

```python
x_train_scaled_tensor.shape
```

```python
# defining model

class MysimpleNN():
    def __init__(self,x):  # for weight and bias
        self.weights = torch.rand(x.shape[1],1 , dtype=torch.float64, requires_grad=True)
        self.bias = torch.zeros(1 , dtype=torch.float64 , requires_grad=True)
    
    def forward(self,x):
        z = torch.matmul(x,self.weights) + self.bias
        y_pred = torch.sigmoid(z)
        return y_pred
    
    def loss_function(self,y_pred,y):
        #clump
        epsilon=1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)
        loss = -(y_train_encoded_tensor*torch.log(y_pred) + (1-y_train_encoded_tensor)*torch.log(1-y_pred)).mean()
        return loss
        
```

```python
model = MysimpleNN(x_train_scaled_tensor)
model.weights
```

```python
model.bias
```

#### Parameters

```python
epochs = 25
learning_rate = 0.1
```

#### Training pipeline

```python
# create model 
model = MysimpleNN(x_train_scaled_tensor)

# define loop
for epoch in range(epochs):
    
    # forward pass
    y_pred = model.forward(x_train_scaled_tensor)
    
    # loss calculate
    loss = model.loss_function(y_pred, y_train_encoded_tensor)
    
    # backward pass
    loss.backward()
    
    # parameter update
    with torch.no_grad():
        model.weights -= learning_rate*model.weights.grad
        model.bias -= learning_rate*model.bias.grad
        
    # zero gradient
    model.weights.grad.zero_()
    model.bias.grad.zero_()
    
    print(f"{epoch+1}, loss = {loss.item()}")
```
