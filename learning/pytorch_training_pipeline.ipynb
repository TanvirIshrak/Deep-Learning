```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head(20)
```

```python
df.shape
```

```python
df.drop(columns=['id','Unnamed: 32'], inplace=True)
```

```python
df.head(20)
```

```python
x = df.iloc[: , 1:]
y = df.iloc[: , 0]

x_train, x_test, y_train, y_test  = train_test_split(
    x,y,test_size=0.2, random_state=42
)
```

```python
df.info()
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)
```

```python
type(y_test_encoded)
```

### Numpy array to tensors conversion

```python
x_train_scaled_tensor = torch.from_numpy(x_train_scaled).float()
x_test_scaled_tensor = torch.from_numpy(x_test_scaled).float()
y_train_encoded_tensor = torch.from_numpy(y_train_encoded).float()
y_test_encoded_tensor = torch.from_numpy(y_test_encoded).float()
```

## Dataset and Dataloader

```python
from torch.utils.data import Dataset , DataLoader

class CustomDataSet(Dataset):
    def __init__(self, features , labels):
        self.features = features
        self.labels = labels
        
    def __len__(self):
        return self.features.shape[0]
    
    def __getitem__(self, index):
        return self.features[index], self.labels[index]
    
        
```

```python
train_dataset = CustomDataSet(x_train_scaled_tensor, y_train_encoded_tensor) 
test_dataset = CustomDataSet(x_test_scaled_tensor, y_test_encoded_tensor)
```

```python
len(train_dataset)
```

```python
train_dataset[2]
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=True)
```

```python
import torch.nn as nn
# defining model

class MysimpleNN(nn.Module):
    def __init__(self,num_features):
        super().__init__()
    
        self.linear = nn.Linear(num_features,1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self,features):
        out = self.linear(features)
        out = self.sigmoid(out)
        return out
    
    def loss_function(self,y_pred,y):
        #clump
        epsilon=1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)
        loss = -(y_train_encoded_tensor*torch.log(y_pred) + (1-y_train_encoded_tensor)*torch.log(1-y_pred)).mean()
        return loss
        
```

#### Training pipeline

```python
# create model 
model = MysimpleNN(x_train_scaled_tensor.shape[1])

# optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# define loop
for epoch in range(epochs):
    
    # forward pass
    y_pred = model(x_train_scaled_tensor) # batch gradient descent
    
    # loss calculate
    loss = model.loss_function(y_pred, y_train_encoded_tensor)
    
    # zero gradient
    # model.linear.weight.grad.zero_()
    # model.linear.bias.grad.zero_()
    optimizer.zero_grad()
    
    # backward pass
    loss.backward()
    
    # parameter update
    # with torch.no_grad():
    #     model.linear.weight -= learning_rate*model.linear.weight.grad
    #     model.linear.bias -= learning_rate*model.linear.bias.grad
    optimizer.step()
        
    # # zero gradient
    # model.linear.weight.grad.zero_()
    # model.linear.bias.grad.zero_()
    
    print(f"{epoch+1}, loss = {loss.item()}")
```

```python
# create model 
# mini batch gradiant descent
model = MysimpleNN(x_train_scaled_tensor.shape[1])

# optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# define loop
batch_size = 32
epoch = 25
n_sample = len(x_train_scaled_tensor)
for epoch in range(epochs):
    
    for start_indx in range(0, n_sample, batch_size):
        end_indx = start_indx + batch_size
        # print(f'Start index : {start_indx} , End index: {end_indx}')
        
        x_batch = x_train_scaled_tensor[start_indx : end_indx]
        y_batch = y_train_encoded_tensor[start_indx : end_indx]
        
        # forward pass
        y_pred = model(x_batch)
        loss = model.loss_function(y_pred , y_batch.reshape(-1,1))
        
        #update steps
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    
    print(f"{epoch+1}, loss = {loss.item()}")
```

### Training pipeline V3

```python
# create model 
model = MysimpleNN(x_train_scaled_tensor.shape[1])

# optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# define loop
for epoch in range(epochs):
    
    for batch_feature , batch_label in train_loader:
        
        #forward pass
        y_pred = model(batch_feature)
        
        # loss calc
        loss = model.loss_function(y_pred, batch_label.reshape(-1,1))
        
        # zero grad
        optimizer.zero_grad()

        # updates
        loss.backward()
        optimizer.step()
        
    print(f"{epoch+1}, loss = {loss.item()}")
```
